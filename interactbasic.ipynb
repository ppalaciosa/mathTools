```python
%matplotlib inline
from ipywidgets import interact, interactive, fixed
import ipywidgets as widgets
import matplotlib.pyplot as plt
import numpy as np
```

```python
t = np.linspace(0,10,10000)
plt.plot(t, np.sin(2*t))
plt.show()
```

```python
def f(w):
    plt.plot(t, np.sin(w*t), color='#1f77b4')
    plt.xlabel('t')
    plt.ylabel('Sin[x]')
    plt.grid(True)
    plt.show()
```

```python
interact(f, w=(0,10,0.1))
```
