```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.ndimage
%matplotlib inline
```

```python
# Get current size
fig_size = plt.rcParams["figure.figsize"]

# Set figure width to 9 and height to 6
fig_size[0] = 9
fig_size[1] = 6
plt.rcParams["figure.figsize"] = fig_size
```

```python
cols = ['angle','P0', 'S0', 'P1', 'S1']
data = pd.read_csv('redGR13-1208.csv', index_col = False, sep=',', skiprows=range(0,1), header=None,names=cols)
x = data.angle
y1 = data.P0
y2 = data.S0
y3 = data.P1
y4 = data.S1
```

```python
ax = plt.subplot(111)
ax.set_title(' ')
#markavery to smooth the line
#https://kaleidoscopicdiaries.wordpress.com/2015/08/12/smooth-line-plots-in-matplotlib-with-fewer-datapoints/
ax.plot(x, y1,'s-', label="Ordem 0 - P", markevery=100) 
ax.plot(x, y2,'s-', label="Ordem 0 - S", markevery=10)
ax.plot(x, y3,'o-', label="Ordem 1 - P")
ax.plot(x, y4,'o-', label="Ordem 1 - S")

ax.grid(True)

ax.legend(loc=2, bbox_to_anchor=(1, 1), fancybox=True, ncol=1)

ax.set_xlabel(r'$\theta _{in}$', size=25)
ax.set_ylabel(r'PotÃªncia Refletida ($\% $)', size=18)

plt.savefig('plot.svg', format='svg',bbox_inches='tight')

plt.show()
```
